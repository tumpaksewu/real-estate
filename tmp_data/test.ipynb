```python
def convert_price(n):
    x = n.split()
    if '$' in x[1]:
        return float(x[0])*90
    if '€' in n.split()[1]:
        return float(x[0])*100
    else:
        return float(x[0])

temp = df['price'].map(convert_price)
```

```python
temp.sort_values(ascending=False)
```

```python
df['price_cleaned'] = temp
```

### Конвертировать метро в станцию + дистанцию

```python
def get_mins_from_metro(x):
    try:
        mins = x.split('(')[1].split(' ')[0]
        return int(mins)
    except:
        return np.nan

def fix_metro(x):
    if isinstance(x, str):
        return x.split(' (')[0]
    return x

df['mins_from_metro'] = df['metro'].map(get_mins_from_metro)
df['metro'] = df['metro'].map(fix_metro)
```

```python
df.mins_from_metro.sort_values(ascending=False)
```

```python
metro_pricing = df.groupby('metro')[['price_cleaned']].median()
```

```python
metro_pricing = metro_pricing.sort_values(by='price_cleaned', ascending=False).head(30)
```

### Суммируем туалеты

```python
import re

def count_toilets(x):
    if isinstance(x, str):
        numbers = re.findall(r'\d+', x)
        return sum(map(int, numbers)) if numbers else np.nan
    return np.nan


df['toilet_count'] = df['toilets'].map(count_toilets)
```
